```python
import numpy as np
import matplotlib.pyplot as plt

# Paso 1: Crear un vector de tamaño 720 con valores aleatorios
vector = np.random.rand(720)

# Paso 2: Hacer reshape a una matriz de dimensiones (120, 6)
matriz = vector.reshape(120, 6)

# Paso 3: Obtener la transpuesta de la matriz
matriz_transpuesta = matriz.T

# Crear dos copias de la matriz transpuesta, una en orden 'F' y otra en orden 'C'
matriz_F = np.array(matriz_transpuesta, order='F')  # Fortran-style
matriz_C = np.array(matriz_transpuesta, order='C')  # C-style

# Paso 4: Crear los gráficos en subplots con diferentes gráficos

# Crear una figura
fig = plt.figure(figsize=(12, 10))

# Panel 1: Gráfico de línea (plot)
ax1 = fig.add_axes([0.05, 0.55, 0.4, 0.4])  # proporciones grandes
ax1.plot(matriz[0], marker='o', label='Datos Fila 1')  # Usando la primera fila de la matriz
ax1.set_title('Gráfico de Línea (Fila 1)')
ax1.set_xlabel('Índice')
ax1.set_ylabel('Valor')
ax1.grid(True)
ax1.legend()

# Panel 2: Gráfico de dispersión (scatter)
ax2 = fig.add_axes([0.55, 0.55, 0.4, 0.4])  # proporciones grandes
ax2.scatter(range(len(matriz[1])), matriz[1], color='r', label='Datos Fila 2')  # Usando la segunda fila
ax2.set_title('Gráfico de Dispersión (Fila 2)')
ax2.set_xlabel('Índice')
ax2.set_ylabel('Valor')
ax2.grid(True)
ax2.legend()

# Panel 3: Gráfico de barras (bar)
ax3 = fig.add_axes([0.05, 0.35, 0.25, 0.15])  # proporciones medianas
ax3.bar(range(len(matriz[2])), matriz[2], color='g', label='Datos Fila 3')  # Usando la tercera fila
ax3.set_title('Gráfico de Barras (Fila 3)')
ax3.set_xlabel('Índice')
ax3.set_ylabel('Valor')
ax3.grid(True)
ax3.legend()

# Panel 4: Histograma
ax4 = fig.add_axes([0.35, 0.35, 0.25, 0.15])  # proporciones medianas
ax4.hist(matriz[3], bins=5, color='b', alpha=0.7, label='Datos Fila 4')  # Usando la cuarta fila
ax4.set_title('Histograma (Fila 4)')
ax4.set_xlabel('Valor')
ax4.set_ylabel('Frecuencia')
ax4.grid(True)
ax4.legend()

# Panel 5: Gráfico circular (Pie chart)
ax5 = fig.add_axes([0.65, 0.35, 0.3, 0.15])  # proporciones más pequeñas
ax5.pie(matriz[4], labels=[f'Seg {i+1}' for i in range(len(matriz[4]))], autopct='%1.1f%%', colors=plt.cm.viridis(np.linspace(0, 1, len(matriz[4]))))
ax5.set_title('Gráfico Circular (Fila 5)')

# Panel 6: Gráfico de área (Stack plot)
ax6 = fig.add_axes([0.05, 0.05, 0.9, 0.2])  # panel grande a lo largo
ax6.stackplot(range(len(matriz[5])), matriz[5], color='c', alpha=0.8, labels=['Datos Fila 6'])  # Usando la sexta fila
ax6.set_title('Gráfico de Área (Fila 6)')
ax6.set_xlabel('Índice')
ax6.set_ylabel('Valor')
ax6.grid(True)
ax6.legend()


plt.show()



```
